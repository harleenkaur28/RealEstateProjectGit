# Missing Attributes Filling

```python
median= housing["RM"].median()
housing["RM"].fillna(median)
#Filling the missing values in the RM column with the median of values
```

```python
from sklearn.impute import SimpleImputer
imputer= SimpleImputer(strategy= "median")
imputer.fit(housing)
#Fit the median value in any column with missing values
```

```python
x= imputer.transform(housing)
#Create a DataFrame with Filled-up values
housing_tr= pd.DataFrame(x, columns= housing.columns)

#Pipeline Below already has an Imputer
```

```python
#feature scaling to have attribute values in the same range
```

# Feature Scaling

1. **Min-Max Scaling(Normalisation)**

`(value-min)/(max-min)`

Goal is to make the values lie between 0-1, for acheiving this functionality sklearn provides a class called `MinMaxScaler`


2. **Standardisation**

`(value-mean)/std`

Goal is to make the standard deviation of the data = 1, for acheiving this functionality sklearn provides a class classed `StandardScaler`

# Creating a Pipeline

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler

my_pipeline= Pipeline([
    ('imputer', SimpleImputer(strategy="median")),
    ('std_scaler', StandardScaler())
    
])
#Pipeline already has an Imputer
```

```python
housing_num_tr= my_pipeline.fit_transform(housing)
```

```python
housing_num_tr
```

# Selecting a Desired Model

```python
from sklearn.tree import DecisionTreeRegressor
#model= DecisionTreeRegressor()
#model.fit(housing_num_tr, housing_labels)
```

```python
from sklearn.ensemble import RandomForestRegressor
model= RandomForestRegressor()
model.fit(housing_num_tr, housing_labels)
```

```python
#Predicting
some_data= housing.iloc[:5]
some_labels= housing_labels.iloc[:5]
prepared_data= my_pipeline.transform(some_data)
model.predict(prepared_data)
```

```python
from sklearn.metrics import mean_squared_error
```

```python
#Evaluating the Model
housing_prediction= model.predict(housing_num_tr)

mse= mean_squared_error(housing_prediction, housing_labels)
rmse= np.sqrt(mse)
```

# Using Cross Validation

```python
from sklearn.model_selection import cross_val_score
scores= cross_val_score(model,housing_num_tr, housing_labels, scoring="neg_mean_squared_error", cv=10)
rmse_score= np.sqrt(-scores)
```

```python
rmse_score
```

```python
def print_score(score, file):
    mean=str(score.mean())
    std= str(score.std())
    file.write("Mean: ")
    file.write(mean)
    file.write("\n")
    file.write("Standard Deviation: ")
    file.write(std)
```

```python
with open("Summary.txt","w") as file_var:
    print_score(rmse_score,file_var)
```
